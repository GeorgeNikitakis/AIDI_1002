```python
# Importing nescessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# Importing dataset from folder
file_location = 'D:\\AIDI\\AIDI1002\\ML Group Project\\PROJECT\\dataset_reviews\\1429_1.csv'
review_data = pd.read_csv(file_location, sep= ',', header= 0)
review_data.head()
```

```python
# Cleaning the dataset by removing unnecessary columns
review_clean = review_data.drop(['id', 'asins', 'keys', 'reviews.date', 'reviews.dateAdded', 'reviews.dateSeen', 'reviews.didPurchase', 'reviews.id', 'reviews.sourceURLs', 'reviews.userCity', 'reviews.userProvince', 'reviews.username'],1)
review_clean.head()
```

```python
review_clean.shape
```

```python
# Exporting dataset to folder under new file name
review_clean.to_csv('D:\\AIDI\\AIDI1002\\ML Group Project\\PROJECT\\dataset_reviews\\data_cleaned.csv', index=False)
```

```python
# Splitting the cleaned dataset into training and testing sets; Also dropped the rows with missing values in specified columns
review_train, review_test = train_test_split(review_clean.dropna(subset=['name', 'brand', 'categories', 'reviews.rating', 'reviews.text', 'reviews.title']), test_size = 0.3, random_state = 100)
```
